```python
import os
import nibabel as nib
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader, random_split
from scipy.ndimage import zoom
import torchio as tio
from typing import Optional, Tuple, List

class ImgDataset(Dataset):
    def __init__(self, input_dir: str, 
                 mask_dir: str, 
                 target_size: Tuple[int, int, int] = (128, 128, 128), 
                 transform: Optional[tio.Transform] = None, 
                 visualize: bool = False
                ) -> None:
        self.input_dir = input_dir
        self.mask_dir = mask_dir
        self.transform = transform
        self.target_size = target_size
        self.visualize = visualize
        self.input_files = sorted(os.listdir(input_dir))
        self.mask_files = sorted(os.listdir(mask_dir))
    
    def __len__(self) -> int:
        """Возвращает количество пар (изображение, маска)"""
        return len(self.input_files)
    
    def resize_volume(self, volume: np.ndarray, target_size: Tuple[int, int, int]) -> np.ndarray:
        """Изменяет размер 3D-объёма под заданные target_size"""
        factors = [target / current for target, current in zip(target_size, volume.shape)]
        return zoom(volume, factors, order=1)
    
    def __getitem__(self, idx: int) -> Tuple[torch.Tensor, torch.Tensor]:
        """Возвращает один элемент датасета (input_tensor, mask_tensor)."""
        input_path = os.path.join(self.input_dir, self.input_files[idx])
        mask_path = os.path.join(self.mask_dir, self.mask_files[idx])
        
        input_img = nib.load(input_path).get_fdata()
        mask_img = nib.load(mask_path).get_fdata()

        # масштабирование
        input_img = self.resize_volume(input_img, self.target_size)
        mask_img = self.resize_volume(mask_img, self.target_size)

        # нормализация входного изображения
        input_img = (input_img - np.min(input_img)) / (np.max(input_img) - np.min(input_img) + 1e-8)
        mask_img = mask_img.astype(np.int64)

        # создаём Subject для TorchIO
        subject = tio.Subject(
            image=tio.ScalarImage(tensor=input_img[np.newaxis, ...]),  # добавляем канал
            label=tio.LabelMap(tensor=mask_img[np.newaxis, ...])        # добавляем канал
        )

        # применяем аугментации
        if self.transform:
            subject = self.transform(subject)

        input_img = subject.image.data[0].numpy()  # (D, H, W)
        mask_img = subject.label.data[0].numpy()   # (D, H, W)
    
        input_tensor = torch.tensor(input_img, dtype=torch.float32).unsqueeze(0)  # [1, D, H, W]
        mask_tensor = torch.tensor(mask_img, dtype=torch.long)                    # [D, H, W]

        return input_tensor, mask_tensor

def get_dataloaders(
        input_dir: str,
        mask_dir: str,
        batch_size: int = 4,
        target_size: Tuple[int, int, int] = (128, 128, 128),
        transform: Optional[tio.Transform] = None,
        val_ratio: float = 0.1,
        visualize: bool = False
    ) -> Tuple[DataLoader, DataLoader]:
    """
    Создаёт train и validation даталоадеры

    params:
        input_dir: путь к папке с томограммами
        mask_dir: путь к папке с масками
        batch_size: размер батча
        target_size: размер объёмов (D, H, W)
        transform: трансформации для аугментаций
        val_ratio: доля валидации
        visualize: включить визуализацию
    returns:
        (train_loader, val_loader): DataLoader для обучения и валидации
    """ 
    dataset = ImgDataset(input_dir, mask_dir, target_size=target_size, transform=transform, visualize=visualize) 
    val_size = int(len(dataset) * val_ratio) 
    train_size = len(dataset) - val_size 
    
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size]) 
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True) 
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False) 
    
    print(f"Train samples: {train_size}") 
    print(f"Val samples: {val_size}") 
    
    return train_loader, val_loader
```

# Выбор архитектуры, разбор технологии



Так как при решении задачи мы пользуемся 3D данными (имеющееся GPU позволяет обучать нейросети на 3D), выбирались архитектуры, позволяющие решать задачу многоклассовой классификации на 3D данных. Эксперимент проводился с архитектурой **V-Net** (https://github.com/faustomilletari/VNet/tree/master). На рисунке выше можно увидеть схематичное изображение архитектуры V-Net.

Была изучена статья (https://arxiv.org/pdf/1606.04797) с обзором архитектуры. В статье приводилась сравнительная таблица алгоритмов loss-функций, можно было понять, что лучшим вариантом для loss в V-Net будет использование Dice-based loss.



```python
import torch
import torch.nn as nn

class VNetBlock(nn.Module):
    """
    Базовый строительный блок VNet.
    Состоит из двух 3D-свёрток (Conv3d) с InstanceNorm и ReLU, 
    а также shortcut-соединения (residual connection).
    """
    def __init__(self, in_channels: int, out_channels: int) -> None:
        """
        Params:
            in_channels: количество входных каналов
            out_channels: количество выходных каналов
        """
        super(VNetBlock, self).__init__()
        self.conv1 = nn.Conv3d(in_channels, out_channels, kernel_size=5, padding=2)
        self.norm1 = nn.InstanceNorm3d(out_channels)
        self.relu1 = nn.ReLU(inplace=True)

        self.conv2 = nn.Conv3d(out_channels, out_channels, kernel_size=5, padding=2)
        self.norm2 = nn.InstanceNorm3d(out_channels)
        self.relu2 = nn.ReLU(inplace=True)

        self.shortcut = nn.Sequential()
        if in_channels != out_channels: # если число каналом меняется, согласуем размерности
            self.shortcut = nn.Sequential(
                nn.Conv3d(in_channels, out_channels, kernel_size=1),
                nn.InstanceNorm3d(out_channels)
            )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """
        Прямой проход блока.

        params:
            x: входной тензор [B, C, D, H, W]
        returns:
            out: выходной тензор той же формы, но с другим числом каналов
        """
        out = self.conv1(x)
        out = self.norm1(out)
        out = self.relu1(out)

        out = self.conv2(out)
        out = self.norm2(out)

        out += self.shortcut(x) # сохранение градиентов
        out = self.relu2(out)
        return out

class VNet(nn.Module):
    """
    3D-сегментационная сеть V-Net.
    Используется для сегментации объёмных данных (например, КТ позвоночника).
    """
    def __init__(self, in_channels: int = 1, out_channels: int = 26) -> None:
        """
        params:
            in_channels: количество каналов входного объёма
            out_channels: количество классов сегментации (включая фон)
        """
        super(VNet, self).__init__()
        self.encoder1 = VNetBlock(in_channels, 16) # сжатие признаков
        self.encoder2 = VNetBlock(16, 32)
        self.encoder3 = VNetBlock(32, 64)
        self.encoder4 = VNetBlock(64, 128)
        
        self.pool = nn.MaxPool3d(2) # делаем после каждого блока, уменьшаем размер в 2 раза по D, H, W
        
        self.middle = VNetBlock(128, 256) # самый глубокий слой
        
        self.upconv4 = nn.ConvTranspose3d(256, 128, kernel_size=2, stride=2) # начало восстановления размера
        self.decoder4 = VNetBlock(256, 128)
        
        self.upconv3 = nn.ConvTranspose3d(128, 64, kernel_size=2, stride=2)
        self.decoder3 = VNetBlock(128, 64)
        
        self.upconv2 = nn.ConvTranspose3d(64, 32, kernel_size=2, stride=2)
        self.decoder2 = VNetBlock(64, 32)
        
        self.upconv1 = nn.ConvTranspose3d(32, 16, kernel_size=2, stride=2)
        self.decoder1 = VNetBlock(32, 16)
        
        self.final = nn.Conv3d(16, out_channels, kernel_size=1) # для нашей задачи out_channels = 25 + 1 (шум)
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        enc1 = self.encoder1(x)
        enc2 = self.pool(enc1)
        enc2 = self.encoder2(enc2)
        enc3 = self.pool(enc2)
        enc3 = self.encoder3(enc3)
        enc4 = self.pool(enc3)
        enc4 = self.encoder4(enc4)
        
        middle = self.pool(enc4)
        middle = self.middle(middle)
        
        dec4 = self.upconv4(middle)
        dec4 = torch.cat([dec4, enc4], dim=1)
        dec4 = self.decoder4(dec4)
        
        dec3 = self.upconv3(dec4)
        dec3 = torch.cat([dec3, enc3], dim=1)
        dec3 = self.decoder3(dec3)
        
        dec2 = self.upconv2(dec3)
        dec2 = torch.cat([dec2, enc2], dim=1)
        dec2 = self.decoder2(dec2)
        
        dec1 = self.upconv1(dec2)
        dec1 = torch.cat([dec1, enc1], dim=1)
        dec1 = self.decoder1(dec1)
        
        out = self.final(dec1) # [B, out_channels, D, H, W]
        return out
```

```python
class MultiClassDiceLoss(nn.Module):
    """
    Классическая реализация Dice Loss для многоклассовой 3D-сегментации.
    Используется для оценки перекрытия между предсказанием модели и разметкой.
    """
    def __init__(self, smooth: float = 1e-5) -> None:
        super(MultiClassDiceLoss, self).__init__()
        self.smooth = smooth # добавляется к числителю и знаменателю чтобы избежать деления на ноль

    def forward(self, preds: torch.Tensor, targets: torch.Tensor) -> torch.Tensor:
        """
        Вычисление Dice Loss.

        params:
            preds: предсказания модели [B, C, D, H, W]
                — B: размер батча
                — C: количество классов
                — D, H, W: размеры 3D-объёма
            targets: истинные метки классов [B, D, H, W]

        returns:
            dice_loss: скалярное значение Dice loss
        """
        num_classes = preds.shape[1]
        preds = F.softmax(preds, dim=1) # преобразование в вероятности классов

        # One-hot кодирование масок [B, D, H, W] → [B, D, H, W, C]
        targets_onehot = F.one_hot(targets, num_classes=num_classes)  # Класс 0 → [1,0,0,...]
        targets_onehot = targets_onehot.permute(0, 4, 1, 2, 3).float() # Переставляем оси: теперь [B, C, D, H, W]

        dims = (0, 2, 3, 4)
        intersection = torch.sum(preds * targets_onehot, dims) # суммируем по всем измерениям где прогноз верный
        union = torch.sum(preds, dims) + torch.sum(targets_onehot, dims)
        dice = (2. * intersection + self.smooth) / (union + self.smooth)

        # Усредняем по всем классам, вычитаем из единицы, чтобы минимизировать, dice_loss = 1 - худший результат
        dice_loss = 1 - dice.mean()
        return dice_loss


class DiceCELoss(nn.Module):
    """
    Комбинированный лосс: Dice Loss + Cross Entropy.
    """
    def __init__(self, smooth: float = 1e-5, dice_weight: float = 0.7, ce_weight: float = 0.3) -> None:
        """
        params:
            smooth (float): сглаживающий коэффициент для DiceLoss.
            dice_weight (float): вес Dice Loss в итоговой функции.
            ce_weight (float): вес CrossEntropy Loss в итоговой функции.
        """
        super(DiceCELoss, self).__init__()
        self.dice = MultiClassDiceLoss(smooth) # компонент Dice
        self.ce = nn.CrossEntropyLoss() # классическая CrossEntropy
        self.dice_weight = dice_weight # Вес Dice
        self.ce_weight = ce_weight # Вес CrossEntropy

    def forward(self, preds: torch.Tensor, targets: torch.Tensor) -> torch.Tensor:
        """
        Вычисление комбинированного Dice + CE Loss.

        params:
            preds: выходы модели [B, C, D, H, W]
            targets: метки классов [B, D, H, W]

        returns:
            total_loss: итоговое значение комбинированного лосса
        """
        dice_loss = self.dice(preds, targets)
        ce_loss = self.ce(preds, targets)
        total_loss = self.dice_weight * dice_loss + self.ce_weight * ce_loss
        return total_loss
```

# Процесс обучения модели V-Net и оценка качества на валидации

```python
def dice_score(preds, targets, num_classes):
    preds = F.one_hot(preds, num_classes).permute(0, 4, 1, 2, 3)
    targets = F.one_hot(targets, num_classes).permute(0, 4, 1, 2, 3)
    preds = preds.float()
    targets = targets.float()

    dims = (0, 2, 3, 4)
    intersection = torch.sum(preds * targets, dims)
    union = torch.sum(preds, dims) + torch.sum(targets, dims)

    dice = (2 * intersection + 1e-5) / (union + 1e-5)
    dice = dice[1:]  # исключаем фон
    return dice.mean().item()
```

```python
from IPython.display import clear_output
import matplotlib.pyplot as plt

def train(model, criterion, optimizer, scheduler, train_loader, val_loader, num_epochs=10, device='cuda'):
    train_losses = []
    val_ious = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0

        for inputs, masks in tqdm.tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]"):
            inputs, masks = inputs.to(device), masks.long().to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, masks)

            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

        epoch_loss = running_loss / len(train_loader.dataset)
        train_losses.append(epoch_loss)

        model.eval()
        dice_scores = []
        
        with torch.no_grad():
            for inputs, masks in tqdm.tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Val]"):
                inputs, masks = inputs.to(device), masks.long().to(device)
                outputs = model(inputs)
                preds = torch.argmax(outputs, dim=1)
                dice = dice_score(preds, masks, num_classes=outputs.shape[1])
                dice_scores.append(dice)
        
        mean_dice = sum(dice_scores) / len(dice_scores)
        val_ious.append(mean_dice)
        
        scheduler.step(mean_dice)

        clear_output(wait=True)
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        
        axes[0].plot(train_losses, label='Train Loss', color='red')
        axes[0].set_xlabel('Epoch')
        axes[0].set_ylabel('Loss')
        axes[0].set_title('Training Loss')
        axes[0].grid(True)
        axes[0].legend()
        
        axes[1].plot(val_ious, label='Validation Dice', color='green')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('Dice')
        axes[1].set_title('Validation Dice')
        axes[1].grid(True)
        axes[1].legend()
        
        plt.tight_layout()
        plt.show()

        print(f"Epoch [{epoch+1}/{num_epochs}]")
        print(f"Train Loss: {epoch_loss:.4f} | Val Dice: {mean_dice:.4f}")

    return train_losses, val_ious
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau

model = VNet(in_channels=1, out_channels=26).cuda()

criterion = DiceCELoss(smooth=1e-5, dice_weight=0.7, ce_weight=0.3)
optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1e-5)

scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=3)

train(model=model, criterion=criterion, optimizer=optimizer, scheduler=scheduler, 
     train_loader=train_loader, val_loader=val_loader, num_epochs=40)
```

```python
def visualize_spine_predictions(model, val_loader, device='cuda', num_samples=3):
    """
    Визуализирует результаты 3D-сегментации позвоночника и вычисляет Dice по каждому классу.

    Эта функция выполняет:
        1. Прогон модели по валидационным данным.
        2. Подсчёт средних Dice по каждому позвонку.
        3. Выводит исходное изображение, предсказанная и истинная маска.

    params:
        model (torch.nn.Module): обученная 3D-сегментационная модель (например, VNet)
        val_loader (torch.utils.data.DataLoader): DataLoader с валидационными 3D-объёмами
        device (str): 'cuda' или 'cpu'
        num_samples (int): количество примеров для визуализации

    returns:
        mean_dice_per_class: средний Dice по каждому классу (включая фон, если он есть)
    """
    model.eval()
    model.to(device)
    samples = [] # сюда будут сохраняться (изображение, предсказание, маска)
    dice_per_class = [] # накопленные по всем батчам Dice

    with torch.no_grad():
        for inputs, masks in val_loader:
            inputs, masks = inputs.to(device), masks.to(device)
            outputs = model(inputs)

            # получаем метку класса с максимальной вероятностью на каждом вокселе
            preds = torch.argmax(outputs, dim=1)  # [B, D, H, W]
            
            # считаем Dice по каждому классу (кроме фона)
            num_classes = outputs.shape[1]
            preds_onehot = F.one_hot(preds, num_classes=num_classes).permute(0, 4, 1, 2, 3).float()
            masks_onehot = F.one_hot(masks, num_classes=num_classes).permute(0, 4, 1, 2, 3).float()

            dims = (0, 2, 3, 4)
            intersection = torch.sum(preds_onehot * masks_onehot, dims)
            union = torch.sum(preds_onehot, dims) + torch.sum(masks_onehot, dims)
            dice = (2. * intersection + 1e-5) / (union + 1e-5)
            dice_per_class.append(dice.cpu().numpy())
            
            for b in range(inputs.shape[0]):
                samples.append((inputs[b].cpu(), preds[b].cpu(), masks[b].cpu()))
            
            if len(samples) >= num_samples: # если необходимое кол-во примеров собрали -> вызод из цикла
                break

    # усредняем Dice по всем батчам
    dice_per_class = np.stack(dice_per_class)
    mean_dice_per_class = dice_per_class.mean(axis=0)

    # выводим метрики по каждому классу
    print("Средний Dice по каждому позвонку (классу):")
    for i, score in enumerate(mean_dice_per_class):
        print(f"Позвонок {i}: {score:.3f}")

    # визуализация
    for i, (img, pred, gt) in enumerate(samples[:num_samples]):
        slice_idx = img.shape[1] // 2  # средний срез по глубине
        img_slice = img[0, slice_idx].numpy()
        pred_slice = pred[slice_idx].numpy()
        gt_slice = gt[slice_idx].numpy()

        fig, axes = plt.subplots(1, 3, figsize=(15, 5)) # исходное изображение, предсказание, истина
        axes[0].imshow(img_slice, cmap='gray')
        axes[0].set_title('Исходное изображение')

        axes[1].imshow(img_slice, cmap='gray')
        axes[1].imshow(pred_slice, alpha=0.5)
        axes[1].set_title('Предсказанная сегментация')

        axes[2].imshow(img_slice, cmap='gray')
        axes[2].imshow(gt_slice, alpha=0.5)
        axes[2].set_title('Истинная разметка')

        plt.suptitle(f"Пример {i+1}")
        plt.show()

    return mean_dice_per_class
```

```python
def dice_score_preds(preds, targets, num_classes):
    """
    preds: [D, H, W] или [B, D, H, W] — предсказанные классы
    targets: [D, H, W] или [B, D, H, W] — истинные классы
    """
    if preds.dim() == 3:
        preds = preds.unsqueeze(0)
        targets = targets.unsqueeze(0)

    preds_onehot = F.one_hot(preds, num_classes=num_classes).permute(0, 4, 1, 2, 3).float()
    targets_onehot = F.one_hot(targets, num_classes=num_classes).permute(0, 4, 1, 2, 3).float()

    dims = (0, 2, 3, 4)
    intersection = torch.sum(preds_onehot * targets_onehot, dims)
    union = torch.sum(preds_onehot, dims) + torch.sum(targets_onehot, dims)
    dice = (2. * intersection + 1e-5) / (union + 1e-5)

    return dice.mean().item()
```
